```python
data["Height"] = np.random.uniform(150,180, size = len(data))
data
```

```python
x = data["Number"].mean()
data["Number"].fillna(x,inplace = True)
data
```

```python
x = data["Age"].mean()
data["Age"].fillna(x,inplace = True)
data
```

```python
x = data["Weight"].mean()
data["Weight"].fillna(x, inplace = True)
data
```

```python
x = data["Salary"].mean()
data["Salary"].fillna(x, inplace = True)
data
```

```python
data.drop_duplicates(inplace = True)
data
```

```python
data.dropna(subset=["College"], inplace=True)
data
```

```python
unique_team = data["Team"].unique()
print(unique_team)
```

```python
team_count = data["Team"].value_counts()
print(team_count)
```

```python
tot_emp = len(data)
print(tot_emp)
```

```python
percentage = (team_count/tot_emp)*100
print(percentage)
```

```python
position_group = data.groupby("Position")["Name"].apply((list))
for position,names in position_group.items():
    print(f"Employees in positions{position}:")
    for name in names:
        print(name)
        print
```

```python
age_group=data["Age"].value_counts()
print(age_group)
```

```python
age_grp = age_group.idxmax()
print("age group of most employees worked is",age_grp)
```

```python
team_position_salary = data.groupby(["Team","Position"])["Salary"].sum()

max_spend_team_position = team_position_salary.idxmax()
max_spend_salary = team_position_salary.max()

print(f"higest spending team and position")
print(f"highest spending team : {max_spend_team_position[0]}")
print(f"highest spending position : {max_spend_team_position[1]}")
print(f"salary : {max_spend_salary}")
```

```python
age = data["Age"]
salary = data["Salary"]

correlation = age.corr(salary)
print(f"correlation between age and salary:{correlation}")

plt.figure(figsize=(10,10))
plt.scatter(age,salary)
plt.title("CORRELATION BETWEEN AGE AND SALARY")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.show()
```
